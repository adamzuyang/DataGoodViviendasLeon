```
map1 = folium.Map(location=(12.454, -86.9986), zoom_start=11)
shape = gpd.read_file('All_Locations_and_roads-line.shp')
shape.to_file('All_Locations_and_roads-line.geojson', driver='GeoJSON')

shape2 = gpd.read_file('Posos-point.shp')
shape2.to_file('Posos-point.geojson', driver='GeoJSON')


shape3 = gpd.read_file('Sutiaba-polygon.shp')
shape3.to_file('Sutiaba-polygon.geojson', driver='GeoJSON')


shape4 = gpd.read_file('Reforestation_sites-polygon.shp')
shape4.to_file('Reforestation_sites-polygon.geojson', driver='GeoJSON')


shape5 = gpd.read_file('Farms_Huertos_Troilo-point.shp')
shape5.to_file('Farms_Huertos_Troilo-point.geojson', driver='GeoJSON')


shape6 = gpd.read_file('Coordinates_of_planting_area-point.shp')
shape6.to_file('Coordinates_of_planting_area-point.geojson', driver='GeoJSON')

shape7 = gpd.read_file('Borders_Comarcas_and_rivers-polygon.shp')
shape7.to_file(' Borders_Comarcas_and_rivers-polygon.geojson', driver='GeoJSON')

folium.GeoJson(
   shape,
).add_to(map1)

# folium.GeoJson(
#    shape2,
# ).add_to(map1)

folium.GeoJson(
   shape3,
).add_to(map1)

folium.GeoJson(
   shape4,
).add_to(map1)

# folium.GeoJson(
#    shape5,
# ).add_to(map1)

# folium.GeoJson(
#    shape6,
# ).add_to(map1)

folium.GeoJson(
   shape7,
).add_to(map1)

shapey1 = pygeoj.load("All_Locations_and_roads-polygon.geojson")
folium.GeoJson(
   shapey1,
).add_to(map1)

shapey = pygeoj.load("nic_admbnda_adm2_gadm_20200813.geojson")
folium.GeoJson(
   shapey,
).add_to(map1)

shapey2 = gpd.read_file('dn742gm4741.shp')
shapey2.to_file('dn742gm4741.geojson', driver='GeoJSON')

folium.GeoJson(
   shapey2,
).add_to(map1)


shape2['lon'] = shape2['geometry'].x
shape2['lat'] = shape2['geometry'].y
lat = list(shape2['lat'])
lon = list(shape2['lon'])
name=list(shape2["Name"])

for lt,ln,nm in zip(lat,lon,name):
    icon = folium.Icon(color="blue", icon="tint")   
    popup = f"""
            {nm} <br>
        """
    popup = folium.Popup(popup, max_width=500)
    folium.Marker(location=[lt,ln],popup=popup, icon= icon).add_to(map1)   


shape5['lon'] = shape5['geometry'].x
shape5['lat'] = shape5['geometry'].y    
lat2 = list(shape5['lat'])
lon2 = list(shape5['lon'])
name2=list(shape5["Name"])


for lt2,ln2,nm2 in zip(lat2,lon2,name2):
    icon = folium.Icon(color="green", icon="leaf")   
    popup = f"""
            {nm2} <br>
        """
    popup = folium.Popup(popup, max_width=500)
    folium.Marker(location=[lt2,ln2],popup=popup, icon= icon).add_to(map1)   
    
    

folium.LayerControl().add_to(map1)

#map1.save('DataGood_Nicaragua_02.html') 
map1
```
