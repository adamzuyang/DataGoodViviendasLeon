```python
shape5 = gpd.read_file('Farms_Huertos_Troilo-point.shp')
shape5.to_file('Farms_Huertos_Troilo-point.geojson', driver='GeoJSON')

shape2 = gpd.read_file('Posos-point.shp')
shape2.to_file('Posos-point.geojson', driver='GeoJSON')


map1 = folium.Map(location=(12.454, -86.9986), zoom_start=11)
lat2 = list(shape5['lat'])
lon2 = list(shape5['lon'])
name2=list(shape5["Name"])

shape2['lon'] = shape2['geometry'].x
shape2['lat'] = shape2['geometry'].y
lat = list(shape2['lat'])
lon = list(shape2['lon'])
name=list(shape2["Name"])

for lt,ln,nm in zip(lat,lon,name):
    icon = folium.Icon(color="blue", icon="tint")   
    popup = f"""
            {nm} <br>
        """
    popup = folium.Popup(popup, max_width=500)
    folium.Marker(location=[lt,ln],popup=popup, icon= icon).add_to(map1)   

for lt2,ln2,nm2 in zip(lat2,lon2,name2):
    icon = folium.Icon(color="green", icon="leaf")   
    popup = f"""
            {nm2} <br>
        """
    popup = folium.Popup(popup, max_width=500)
    folium.Marker(location=[lt2,ln2],popup=popup, icon= icon).add_to(map1)   
    
    

folium.LayerControl().add_to(map1)

#map1.save('DataGood_Nicaragua_02.html') 
map1
```
